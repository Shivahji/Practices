```

import numpy as np
import matplotlib.pyplot as plt

# تعریف x با فاصله حدودی 0.01
x = np.linspace(-1, 1, 100)

# ===== سوال 1 =====
y1 = np.sin(x)

plt.plot(x, y1, color='blue', linewidth=2, label='y = sin(x)')
plt.title("نمودار تابع y = sin(x)")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.savefig("soal1_100dpi.png", dpi=100)
plt.savefig("soal1_300dpi.png", dpi=300)
plt.clf()

# ===== سوال 2 =====
y2 = 2 * np.sin(x / 2) * (np.cos(x / 4)**2 - np.sin(x / 4)**2)

plt.plot(x, y2, color='red', linewidth=2, label='تابع ترکیبی')
plt.title("نمودار تابع ترکیبی")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.savefig("soal2_100dpi.png", dpi=100)
plt.savefig("soal2_300dpi.png", dpi=300)
plt.clf()

# ===== سوال 3 =====
plt.plot(x, y1, color='blue', linewidth=2, label='y = sin(x)')
plt.plot(x, y2, color='red', linewidth=2, label='تابع ترکیبی')
plt.title("ترکیب دو تابع")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.savefig("soal3_100dpi.png", dpi=100)
plt.savefig("soal3_300dpi.png", dpi=300)
plt.clf()

print("تمام نمودارها برای سه سوال تمرین ۴ با موفقیت ذخیره شدند.")
```
