```python
from scipy.io import wavfile
import numpy as np
import matplotlib.pyplot as plt

# 1. خواندن فایل صوتی
file_name = "sample.wav"  
sample_rate, signal = wavfile.read(file_name) 

# 2. تبدیل به مونو اگر استریو است
if len(signal.shape) > 1:
    signal = signal.mean(axis=1)

# 3. تحلیل FFT
fft_result = np.fft.fft(signal)
magnitude = np.abs(fft_result)
frequencies = np.fft.fftfreq(len(signal), 1/sample_rate)

# 4. نمایش نتایج
dominant_freq = frequencies[np.argmax(magnitude)]
print(f"فرکانس غالب: {abs(dominant_freq):.2f} Hz")
print(f"نرخ نمونه‌برداری: {sample_rate} Hz")

# 5. رسم نمودار
plt.figure(figsize=(12, 4))
plt.plot(frequencies[:len(frequencies)//2], magnitude[:len(magnitude)//2])
plt.title("طیف فرکانسی")
plt.xlabel("فرکانس (Hz)")
plt.grid()
plt.show()
```

```python
import numpy as np
from scipy.io import wavfile
import matplotlib.pyplot as plt

# خواندن فایل صوتی
file_name = "Shiva_sound.wav" 
sample_rate, signal = wavfile.read(file_name)

# نمایش نرخ نمونه‌برداری
print(f"نرخ نمونه‌برداری (Hz): {sample_rate}")

# اگر سیگنال استریو است، به مونو تبدیل کنید
if len(signal.shape) > 1:
    signal = signal.mean(axis=1)

# حذف مؤلفه DC (میانگین سیگنال)
signal = signal - np.mean(signal)

# محاسبه FFT
fft_result = np.fft.fft(signal)
magnitude = np.abs(fft_result)
frequencies = np.fft.fftfreq(len(signal), 1 / sample_rate)

# فقط فرکانس‌های مثبت را بررسی کنید
positive_mask = frequencies >= 0
positive_freq = frequencies[positive_mask]
positive_magnitude = magnitude[positive_mask]

# یافتن فرکانس غالب (با رد کردن فرکانس ۰ هرتز)
dominant_idx = np.argmax(positive_magnitude[1:]) + 1  # +1 برای جلوگیری از انتخاب فرکانس ۰
dominant_freq = positive_freq[dominant_idx]

print(f"فرکانس غالب (Hz): {dominant_freq:.2f}")

# رسم نمودار فرکانسی
plt.figure(figsize=(12, 4))
plt.plot(positive_freq[:len(positive_freq)//2], positive_magnitude[:len(positive_freq)//2])
plt.title("طیف فرکانسی")
plt.xlabel("فرکانس (Hz)")
plt.ylabel("دامنه")
plt.grid()
plt.show()
```

چگونه فرکانس غالب را به ۳۰۰۰ هرتز رساندید؟ (با تغییر تن صدا ، استفاده از نرم‌افزارهای ویرایش صوت).  
     - آیا این فرکانس با صدای طبیعی انسان قابل دستیابی است؟ (پاسخ: بله، با صدای **بسیار زیر**، اما معمولاً نیاز به ابزارهای پردازش صوت دارد).  

     فرکانس‌های بالا (مثل 3000 هرتز) معمولاً با **صداهای زیر** (مثل سوت، جیغ، یا حروف "س" و "ف") تولید می‌شوند.
