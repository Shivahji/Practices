```python
import matplotlib.pyplot as plt
import numpy as np

def create_bar_chart():
    # داده‌ها برای نمودار میله‌ای
    categories = ['بازه اول (0-5)', 'بازه دوم (5-10)', 'بازه سوم (10-15)', 'بازه چهارم (15-19.7)']
    counts = [2, 3, 3, 3]  # تعداد مفاهیم در هر بازه
    
    # ایجاد نمودار میله‌ای
    plt.figure(figsize=(10, 6))
    bars = plt.bar(categories, counts, color=['skyblue', 'lightgreen', 'salmon', 'gold'])
    
    # عنوان و برچسب‌ها
    plt.title('توزیع مفاهیم در بازه‌های زمانی مختلف', fontsize=14)
    plt.xlabel('بازه زمانی', fontsize=12)
    plt.ylabel('تعداد مفاهیم', fontsize=12)
    
    # نمایش تعداد روی هر میله
    for bar in bars:
        height = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2., height,
                 '%d' % int(height),
                 ha='center', va='bottom')
    
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.savefig('bar_chart.png', dpi=300, bbox_inches='tight')
    plt.show()

def create_pie_chart():
    # داده‌ها برای نمودار دایره‌ای
    labels = ['بازه اول (0-5)', 'بازه دوم (5-10)', 'بازه سوم (10-15)', 'بازه چهارم (15-19.7)']
    sizes = [2, 3, 3, 3]  # تعداد مفاهیم در هر بازه
    colors = ['skyblue', 'lightgreen', 'salmon', 'gold']
    
    # ایجاد نمودار دایره‌ای
    plt.figure(figsize=(8, 8))
    patches, texts, autotexts = plt.pie(sizes, labels=labels, colors=colors,
                                        autopct='%1.1f%%', startangle=90,
                                        textprops={'fontsize': 12})
    
    # عنوان
    plt.title('درصد توزیع مفاهیم در بازه‌های زمانی', fontsize=14)
    
    # نمایش درصدها با فونت بزرگتر
    for autotext in autotexts:
        autotext.set_fontsize(12)
    
    plt.axis('equal')  # دایره کامل
    plt.tight_layout()
    plt.savefig('pie_chart.png', dpi=300, bbox_inches='tight')
    plt.show()

# اجرای توابع برای ایجاد نمودارها
create_bar_chart()
create_pie_chart()

###Shiva_hajalz
```
